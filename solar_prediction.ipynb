```python
from google.colab import files
uploaded = files.upload()
import zipfile
import os

zip_filename = list(uploaded.keys())[0]
extract_path = "/content/solar_data"

with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(" Files extracted to:", extract_path)
print(os.listdir(extract_path))
import pandas as pd

gen_data = pd.read_csv(f"{extract_path}/Plant_1_Generation_Data.csv")
weather_data = pd.read_csv(f"{extract_path}/Plant_1_Weather_Sensor_Data.csv")

print("\n Loaded Generation Data:", gen_data.shape)
print("Loaded Weather Data:", weather_data.shape)
print("\nüßπ Cleaning data...")

gen_data['DATE_TIME'] = pd.to_datetime(gen_data['DATE_TIME'], errors='coerce', dayfirst=True)
weather_data['DATE_TIME'] = pd.to_datetime(weather_data['DATE_TIME'], errors='coerce')
gen_data.dropna(subset=['DATE_TIME'], inplace=True)
weather_data.dropna(subset=['DATE_TIME'], inplace=True)
gen_data.drop_duplicates(inplace=True)
weather_data.drop_duplicates(inplace=True)
gen_data.fillna(gen_data.mean(numeric_only=True), inplace=True)
weather_data.fillna(weather_data.mean(numeric_only=True), inplace=True)

print(" Data cleaned successfully!")
print("\nüîó Merging data...")
merged_df = pd.merge(gen_data, weather_data, on=["DATE_TIME", "PLANT_ID"])
print("Merged dataset shape:", merged_df.shape)
for col in ['SOURCE_KEY_x', 'SOURCE_KEY_y']:
    if col in merged_df.columns:
        merged_df.drop(col, axis=1, inplace=True)
merged_df = merged_df[(merged_df['DC_POWER'] >= 0) & (merged_df['AC_POWER'] >= 0)]
clean_path = "/content/cleaned_solar_data.csv"
merged_df.to_csv(clean_path, index=False)
print("\n Cleaned dataset saved at:", clean_path)
print("\nüîç Preview of cleaned dataset:")
merged_df.head()
```
